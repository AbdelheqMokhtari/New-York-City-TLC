```python
import pandas as pd 
```

```python
df = pd.read_csv("DataSet/2017_Yellow_Taxi_Trip_Data.csv")
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
# Convert columns to datetime
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Calculate time difference and create a new column
df['TimeDifference'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']


# Extract hours and minutes
df['Hours'] = df['TimeDifference'].dt.components['hours']
df['Minutes'] = df['TimeDifference'].dt.components['minutes']

# Convert hours to Minutes
df['Minutes'] = df['Minutes'] + df['Hours'] * 60

# Drop the "Hours" ,"Time Difference" columns
df = df.drop(columns=['Hours','TimeDifference'])
```

```python
# Convert 'VendorID ' to categorical data
df['VendorID'] = df['VendorID'].astype('category')

# Convert 'VendorID ' to categorical data
df['RatecodeID'] = df['RatecodeID'].astype('category')

# Convert 'store_and_fwd_flag ' to categorical data
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].astype('category')

# Convert 'VendorID ' to categorical data
df['payment_type'] = df['payment_type'].astype('category')

# Convert 'PULocationID ' to categorical data
df['PULocationID'] = df['PULocationID'].astype('category')

# Convert 'DOLocationID' to categorical data
df['DOLocationID'] = df['DOLocationID'].astype('category')

# Convert 'extra' to categorical data
df['extra'] = df['extra'].astype('category')
```
